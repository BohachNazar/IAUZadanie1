#### Úprava formátu dátumov na jednotný formát:

```python
def normalize_date_series(s):
    s_clean = s.astype(str).str.strip()
    # odstránenie času
    s_clean = s_clean.str.replace(r",\s*\d{2}:\d{2}:\d{2}$", "", regex=True)
    # nahradenie bodiek a medzier
    s_clean = s_clean.str.replace(".", "/", regex=False)
    s_clean = s_clean.str.replace(r"\s+", " ", regex=True)
    
    #viacnásobná konverzia
    out = pd.to_datetime(s_clean, errors='coerce')
    mask = out.isna()
    out2 = pd.to_datetime(s_clean[mask], errors='coerce', dayfirst=True)
    out.loc[mask] = out2
    
    common_formats = [
        "%d %b %Y", 
        "%d %B %Y",      
        "%Y-%m-%d",      
        "%Y/%m/%d",      
        "%d/%m/%Y",      
        "%m/%d/%Y",      
    ]
    mask = out.isna()
    if mask.any():
        for fmt in common_formats:
            try:
                tmp = pd.to_datetime(s_clean[mask], format=fmt, errors='coerce')
                fill_mask = mask.copy()
                fill_mask.loc[:] = False
                fill_mask[mask] = tmp.notna().values
                out.loc[fill_mask] = tmp[tmp.notna()]
                mask = out.isna()
                if not mask.any():
                    break
            except Exception:
                pass
    return out

patient['registration_dt'] = normalize_date_series(patient['registration'])
station['revision_dt'] = normalize_date_series(station['revision'])

print('Neúspešné konverzie po normalizácii - patient:', patient['registration_dt'].isna().sum())
print('Neúspešné konverzie po normalizácii - station:', station['revision_dt'].isna().sum())
```

Vytvorili sme si pomocne stĺpce, kde máme upravené dátumy:

stlpec `registration_dt` v patient

stlpec `revision_dt` v station

###  (B-2b) Kontrola správnosť v dátach

- či obsahujú abnormálne hodnoty

- či obsahujú nelogické dátové vzťahy, ktoré sú následkom dátovej kolekcie a anotovania dát   

```python
# Kontrola správnosti pre všetky numerické stĺpce (okrem geo)
geo_exclude = [c for c in ['latitude', 'longitude'] if c in obs.columns]
key_cols = [c for c in obs.select_dtypes(include=['number']).columns if c not in geo_exclude]

fig, axes = plt.subplots(nrows=int(np.ceil(len(key_cols)/4)), ncols=4, figsize=(16, 3*int(np.ceil(len(key_cols)/4))))
axes = axes.flatten() if hasattr(axes, 'flatten') else [axes]

iqr_rows = []
for i, col in enumerate(key_cols):
    series = obs[col].dropna()
    if i < len(axes):
        axes[i].boxplot(series, vert=True, tick_labels=[col])
        axes[i].grid(True, alpha=0.3)
    # IQR 
    Q1, Q3 = series.quantile(0.25), series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    iqr_rows.append({'col': col, 'Q1': Q1, 'Q3': Q3, 'IQR': IQR, 'Lower': lower, 'Upper': upper, 'Outliers': ((series<lower)|(series>upper)).sum()})

for j in range(len(key_cols), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()

iqr_df = pd.DataFrame(iqr_rows)
print("IQR prehľad (počet bodov mimo whiskers) – všetky numerické stĺpce:")
print(iqr_df.to_string(index=False))
```

V časti 1.1.B sme zistili, že všetky hodnoty patria do definovaných intervalov.

(C-1b) Vychýlené hodnoty (outlier detection), vyskúšajte riešiť problém min. 2 technikami
- odstránenie vychýlených alebo odľahlých pozorovaní
- nahradenie vychýlenej hodnoty hraničnými hodnotami rozdelenia (napr. 5%, 95%)

1) IQR (InterQuartile Range) – odstránenie riadkov s outliermi.
2) Standard deviation (štandardná odchýlka) – 5. a 95. percentil

```python
# Detekcia a odstránenie IQR outlierov (bez geo stĺpcov)
geo_exclude = [c for c in ['latitude', 'longitude'] if c in obs.columns]
num_cols = [c for c in obs.select_dtypes(include=['number']).columns if c not in geo_exclude]
iqr_df = obs.copy()

row_is_outlier = pd.Series(False, index=iqr_df.index)
per_col_counts = {}
for col in num_cols:
    Q1 = iqr_df[col].quantile(0.25)
    Q3 = iqr_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mask = (iqr_df[col] < lower) | (iqr_df[col] > upper)
    per_col_counts[col] = int(mask.sum())
    row_is_outlier |= mask

print("Počty IQR outlierov na stĺpec (top 10):")
print(sorted(per_col_counts.items(), key=lambda x: x[1], reverse=True)[:10])
print(f"Riadky s outlierom aspoň v jednom stĺpci: {int(row_is_outlier.sum())}")
print("Tvar pôvodný:", obs.shape)
obs_iqr = iqr_df[~row_is_outlier].copy()
print("Tvar po odstránení IQR outlierov:", obs_iqr.shape)
```

```python
# C-2b: Technika 2 Standard Deviation (winsorizácia bez geo stĺpcov)
sd_df = obs.copy()
geo_exclude = [c for c in ['latitude', 'longitude'] if c in sd_df.columns]
num_cols = [c for c in sd_df.select_dtypes(include=['number']).columns if c not in geo_exclude]
summary = []
for col in num_cols:
    col_series = sd_df[col]
    mean, std = col_series.mean(), col_series.std()
    if std == 0 or np.isnan(std):
        summary.append({'col': col, 'z_outliers': 0, 'p5': np.nan, 'p95': np.nan})
        continue
    z = np.abs((col_series - mean) / std)
    z_outliers = int((z > 3).sum())
    p5, p95 = col_series.quantile(0.05), col_series.quantile(0.95)
    sd_df[col] = col_series.clip(lower=p5, upper=p95)
    summary.append({'col': col, 'z_outliers': z_outliers, 'p5': p5, 'p95': p95})

print("Z-score outliery (|z|>3) – počty na stĺpec (top 10):")
print(sorted([(s['col'], s['z_outliers']) for s in summary], key=lambda x: x[1], reverse=True)[:10])
print("Tvar po winsorizácii:", sd_df.shape)
obs_sd = sd_df
```

### Porovnanie distribúcií: Pôvodné vs. IQR odstránenie vs. Winsorizácia

Vizuálne porovnanie účinku oboch techník na rozdelenie dát.

```python
# Porovnanie distribúcií pre všetky numerické stĺpce (okrem geo)
geo_exclude = [c for c in ['latitude', 'longitude'] if c in obs.columns]
num_cols = [c for c in obs.select_dtypes(include=['number']).columns if c not in geo_exclude]

n_cols_display = len(num_cols)
fig, axes = plt.subplots(nrows=n_cols_display, ncols=3, figsize=(15, 3*n_cols_display))

for idx, col in enumerate(num_cols):
    # Pôvodné dáta
    axes[idx, 0].hist(obs[col].dropna(), bins=30, color='steelblue', edgecolor='black', alpha=0.7)
    axes[idx, 0].set_title(f'{col} - Pôvodné', fontsize=10)
    axes[idx, 0].set_ylabel('Počet', fontsize=9)
    axes[idx, 0].grid(True, alpha=0.3)
    
    # Po IQR odstránení
    axes[idx, 1].hist(obs_iqr[col].dropna(), bins=30, color='seagreen', edgecolor='black', alpha=0.7)
    axes[idx, 1].set_title(f'{col} - IQR odstránenie', fontsize=10)
    axes[idx, 1].grid(True, alpha=0.3)
    
    # Po winsorizácii
    axes[idx, 2].hist(obs_sd[col].dropna(), bins=30, color='coral', edgecolor='black', alpha=0.7)
    axes[idx, 2].set_title(f'{col} - Winsorizácia', fontsize=10)
    axes[idx, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Porovnanie boxplotov: Pôvodné vs. IQR odstránenie vs. Winsorizácia
Pre každý numerický stĺpec (okrem geo) zobrazíme tri boxploty vedľa seba, aby bolo jasne vidieť účinok čistenia.

```python
geo_exclude = [c for c in ['latitude', 'longitude'] if c in obs.columns]
num_cols = [c for c in obs.select_dtypes(include=['number']).columns if c not in geo_exclude]

for col in num_cols:
    data = [obs[col].dropna(), obs_iqr[col].dropna(), obs_sd[col].dropna()]
    labels = ['Pôvodné', 'IQR odstránenie', 'Winsorizácia']
    minval = min([d.min() for d in data if len(d)>0])
    maxval = max([d.max() for d in data if len(d)>0])
    plt.figure(figsize=(8,2.5))
    bplot = plt.boxplot(data, vert=False, patch_artist=True, tick_labels=labels)
    colors = ['steelblue', 'seagreen', 'coral']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.6)
    plt.title(f'Porovnanie: {col}')
    plt.xlim(minval, maxval)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```
